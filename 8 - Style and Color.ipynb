# <u>Style and Color.

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

---

# <u>Imports.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

---

# <u>Data to plot.

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

---

# <u>Styles.

- ***sns.set_style(): Sets the parameters that control the general style of the plots.***

- ***Parameters for set_style():***

        - 'darkgrid'
        - 'whitegrid'
        - 'dark'
        - 'white'
        - 'ticks'

```python
# Countplot.

sns.countplot(x = 'sex', data = tips)

plt.show()
```

```python
# Shows a darkgrid.

sns.set_style('darkgrid')

sns.countplot(x = 'sex', data = tips)

plt.show()
```

```python
# Shows a whitegrid.
# More preferred.

sns.set_style('whitegrid')

sns.countplot(x = 'sex', data = tips)

plt.show()
```

```python
# Shows a dark background.

sns.set_style('dark')

sns.countplot(x = 'sex', data = tips)

plt.show()
```

```python
# Shows a white background.
# Default style.

sns.set_style('white')

sns.countplot(x = 'sex', data = tips)

plt.show()
```

```python
# Shows ticks on axis.

sns.set_style('ticks')

sns.countplot(x = 'sex', data = tips)

plt.show()
```

```python
# Can also use color argument from Matplotlib.

sns.set_style('ticks')

sns.countplot( x = 'sex', data = tips, color = 'grey')

plt.show()
```

```python
# Passing palette without assigning hue is deprecated and will be removed in v0.14.0.
# Treat x also as hue.

sns.set_style('ticks')

sns.countplot( x = 'sex', data = tips, palette = 'grey', hue = 'sex')

plt.show()
```

---

# <u>Spine Removal.

- ***sns.despine(): Removes the top and right spines from plot(s).***

- Parameters: top, right, left, bottom = boolean

- top, right, left, bottom : boolean, optional
    If True, remove that spine.

```python
# Removes top and right spines.

sns.countplot(x = 'sex', data = tips)

sns.despine()

plt.show()
```

```python
# Removes top and right and left spines.

sns.countplot(x = 'sex', data = tips)

sns.despine(left = True)

plt.show()
```

```python
sns.countplot(x = 'sex', data = tips)

sns.despine(left = True, bottom = True)

plt.show()
```

```python
sns.set_style('ticks')

sns.countplot(x = 'sex', data = tips)

sns.despine(right = False)

plt.show()
```

---

# <u>Size and Aspect.

- We can use matplotlib's **plt.figure(figsize = (width, height)** to change the size of most seaborn plots.

- We can control the size and aspect ratio of most seaborn grid plots by passing in **parameters: height, and aspect.** For example:

```python
# Non Grid Plot

plt.figure(figsize = (12, 3))

sns.set_style('whitegrid')

sns.countplot(x = 'sex', data = tips, hue = 'sex', palette = 'grey')

sns.despine()

plt.show()
```

```python
tips.head()
```

```python
# Grid Type Plots.
# Linear Model plot.

sns.set_style('whitegrid')

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex', palette = 'coolwarm', height = 5, aspect = 0.5)

sns.despine()

plt.show()
```

```python
# Grid Type Plots.
# FacetGrid.

sns.set_style('whitegrid')

g = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex', palette = 'rainbow', height = 5, aspect = 1)

g.map(plt.scatter, 'total_bill', 'tip')

sns.despine()

plt.show()
```

# <u>Scale and Context.

- ***set_context(context=None, font_scale=1): Allows us to override default parameters:***

- context parameter: name of a preconfigured set

            values = 'paper', 'notebook', 'talk', 'poster'

- font_scale: Separate scaling factor to independently scale the size of the font elements(float).

```python
sns.set_context('paper', font_scale = 2)

sns.countplot(x = 'sex', data = tips, hue = 'sex', palette = 'coolwarm')

plt.show()
```

```python
sns.set_context('notebook', font_scale = 2)

sns.countplot(x = 'sex', data = tips, hue = 'sex', palette = 'coolwarm')

plt.show()
```

```python
sns.set_context('talk', font_scale = 2)

sns.countplot(x = 'sex', data = tips, hue = 'sex', palette = 'coolwarm')

plt.show()
```

```python
sns.set_context('poster', font_scale = 2)

sns.countplot(x = 'sex', data = tips, hue = 'sex', palette = 'coolwarm')

plt.show()
```
