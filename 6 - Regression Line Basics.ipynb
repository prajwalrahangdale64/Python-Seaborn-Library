# <u>Regression line basics.

- A regression line is a straight line that best represents the relationship between two variables in a dataset.
- In short: ***It’s the “line of best fit” that shows the overall trend of the data, usually in the form: y = mx + b***
- Example:
    
        - Suppose we plot people’s height on the x-axis and their weight on the y-axis.
        - We’ll see a cloud of scatter points.
        - A regression line is a line drawn through those points that best summarizes the overall trend.
        - If height increases, does weight generally increase too? The regression line shows that trend.

### Slope and intercept.

- Mathematically the linear regression line is: y = mx + b

        where;
            m = slope
            b = intercept
            y = dependent variable
            x = independent variable
- ***Slope(m):***

        - The slope tells us how steep the line is.
        - It shows how much y changes when x changes by 1 unit.
        - Example: If m = 2, then for every increase of 1 in x, y increases by 2.
        - In a graph, it’s literally the “tilt” of the line.
        - Two-point slope formula: exact slope of the line through 2 specific points.
                slope = rise/run or change in y / change in x
                slope = (y2 - y1) / (x2 - x1)
                Note: tanΘ = opposite side / adjacent side 



- Regression slope formula: average slope that best represents the trend across all points (minimizes squared error).



This formula comes from the least squares method.

- ***Intercept(b):***

        - The intercept is where the line crosses the y-axis (when x = 0).
        - It tells us the starting value of y if x were 0.

### What is a residual?

***A residual is the difference between the actual data point and the predicted point on the regression line.***
