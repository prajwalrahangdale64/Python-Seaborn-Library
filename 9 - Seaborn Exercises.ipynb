### 6)

```python
corr_titanic =  titanic.corr(numeric_only = True)
```

```python
corr_titanic.head(10)
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

fig = plt.figure(figsize = (4.5, 4.5))
axes = fig.add_axes([0,0,1,1])

sns.heatmap(data = corr_titanic, annot = True, cmap = 'coolwarm', linecolor = 'white', linewidths = 3)
axes.set_title('titanic.corr()')


plt.show()
```

### 7)

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

facet_grid1 = sns.FacetGrid(data = titanic, col = 'sex', height = 5, aspect = 0.8)

facet_grid1.map(sns.histplot, 'age')

plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

facet_grid1 = sns.FacetGrid(data = titanic, col = 'sex', hue = 'alive', palette = 'rainbow', height = 5, aspect = 0.8)

facet_grid1.map(sns.histplot, 'age').add_legend()

plt.show()
```
